```python
# Linked List (Youtube Tutorial)
class Node:
    def __init__(self, data=None, next=None):
        self.data = data # data it self
        self.next = next
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def insert_at_begining(self, data):
        node = Node(data, self.head) # Create a new data and retrieve the current head (could be none)
        self.head = node # Let the node be the new head
        
    def print(self):
        if self.head is None:
            print('Linked list is empty')
            return
        
        itr = self.head # Let the iterator be the head of the linked list
        llstr = '' # Set a linked list string
        
        while itr: # While the head exist
            llstr += str(itr.data)+'-->' # Modify the string
            itr = itr.next # Move the iterator to the next element
            
        print(llstr)
        
        
ll = LinkedList()
ll.insert_at_begining(5)
ll.insert_at_begining(89)
ll.print()
```

### 203.移除链表元素

题意：删除链表中等于给定值 val 的所有节点。

示例 1：
输入：head = [1,2,6,3,4,5,6], val = 6
输出：[1,2,3,4,5]

示例 2：
输入：head = [], val = 1
输出：[]

示例 3：
输入：head = [7,7,7,7], val = 7
输出：[]

```python
# Sample answer

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution:
    def removeElements(self, head, val):
        dummy_head = ListNode(next=head) #添加一个虚拟节点
        cur = dummy_head #Set the dummy head as the current 节点
        
        while(cur.next!=None): #如果cur.next存在(不存在意味着linked list已经全部被读取)
            if(cur.next.val == val): # 如果
                cur.next = cur.next.next #删除cur.next节点
            else:
                cur = cur.next
        return dummy_head.next
    
# 这段code无法展示出结果，需要define一整个linkedlist才可以
```

### 707. 设计链表

在链表类中实现这些功能：

get(index)：获取链表中第 index 个节点的值。如果索引无效，则返回-1。

addAtHead(val)：在链表的第一个元素之前添加一个值为 val 的节点。插入后，新节点将成为链表的第一个节点。

addAtTail(val)：将值为 val 的节点追加到链表的最后一个元素。

addAtIndex(index,val)：在链表中的第 index 个节点之前添加值为 val  的节点。如果 index 等于链表的长度，则该节点将附加到链表的末尾。如果 index 大于链表长度，则不会插入节点。如果index小于0，则在头部插入节点。

deleteAtIndex(index)：如果索引 index 有效，则删除链表中的第 index 个节点。

```python
# Sample answer
# 单链表
class Node:
    
    def __init__(self, val):
        self.val = val
        self.next = None


class MyLinkedList:

    def __init__(self):
        self._head = Node(0)  # 虚拟头部节点 
        self._count = 0  # 添加的节点数

    def get(self, index: int) -> int:
        """
        Get the value of the index-th node in the linked list. If the index is invalid, return -1.
        """
        if 0 <= index < self._count:
            node = self._head
            for _ in range(index + 1):
                node = node.next
            return node.val
        else:
            return -1
        
    def addAtHead(self, val: int) -> None:
        """
        Add a node of value val before the first element of the linked list. After the insertion, the new node will be the first node of the linked list.
        """
        self.addAtIndex(0, val)

    def addAtTail(self, val: int) -> None:
        """
        Append a node of value val to the last element of the linked list.
        """
        self.addAtIndex(self._count, val)

    
    #     在第 index 个节点之前插入一个新节点，例如index为0，那么新插入的节点为链表的新头节点。
    #     如果 index 等于链表的长度，则说明是新插入的节点为链表的尾结点
    #     如果 index 大于链表的长度，则返回空    
    def addAtIndex(self, index: int, val: int) -> None:
        """
        Add a node of value val before the index-th node in the linked list. If index equals to the length of linked list, the node will be appended to the end of linked list. If index is greater than the length, the node will not be inserted.
        """
        if index < 0:
            index = 0
        elif index > self._count: # If the exceed the length of the linked_list, return nothing.
            return

        # 计数累加
        self._count += 1

        add_node = Node(val)
        prev_node, current_node = None, self._head # Set the current_node to the head node initially
        for _ in range(index + 1): # 从head开始iterate, 然后iterate到index的node (add_node)
            prev_node, current_node = current_node, current_node.next
        else: # 把prev_node的next node指向新的add_node，把add_node的next node指向current node
            prev_node.next, add_node.next = add_node, current_node
    
    
    def deleteAtIndex(self, index: int) -> None:
        """
        Delete the index-th node in the linked list, if the index is valid.
        """
        if 0 <= index < self._count:
            # 计数-1
            self._count -= 1
            prev_node, current_node = None, self._head
            for _ in range(index + 1): # 从head开始iterate, 然后iterate到index的node
                prev_node, current_node = current_node, current_node.next
            else:
                prev_node.next, current_node.next = current_node.next, None
    
    # ----------------------------------------------------------------------------------------------
    # Utility function to visualise the linked list
    def print(self):
        if self._head is None:
            print('Linked list is empty')
            return
        
        itr = self._head # Let the iterator be the head of the linked list
        llstr = '' # Set a linked list string
        
        while itr: # While the head exist
            llstr += str(itr.val)+'-->' # Modify the string
            itr = itr.next # Move the iterator to the next element
            
        print(llstr)
        
    # Test leetcode 203 here    
    def removeElements(self, val):
        # 这个解法里不需要setup dummy_head，因为这个解法里已经设置了一个dummy head 0
        cur = self._head

        while(cur.next!=None): #如果cur.next存在(不存在意味着linked list已经全部被读取)
            if(cur.next.val == val): # 如果
                cur.next = cur.next.next #删除cur.next节点
            else:
                cur = cur.next
        return
    
    # This function finds the length of the linked_list, good to know (This method counts the dummy node as well)
    def find_length(self):
        cur = self._head
        length = 0
        while (cur!=None): # If I do not want to count the dummy node, use while (cur.next != none)
            length+=1
            cur = cur.next
        print(length)
        return length
    
    
```

```python
ll = MyLinkedList()
ll.addAtHead(5)
ll.addAtTail(89)
ll.print()
print(ll.get(1))
print(ll.get(2))
ll.addAtIndex(1,100)
ll.addAtIndex(10,103)
ll.print()
ll.removeElements(100)
ll.removeElements(5)
ll.print()
ll.find_length()
```

### 206.反转链表

题意：反转一个单链表。

示例: 输入: 1->2->3->4->5->NULL 输出: 5->4->3->2->1->NULL

```python
# Sample answer
class Solution:
    def reverseList(self, head):
        # Define the current node and previous node
        cur = head
        pre = None 
        while (cur!=None):
            temp = cur.next
            cur.next = pre # 将下一个node指向previous node
            # 转向后，开始操作下一个node
            pre = cur
            cur = temp
        return pre
        
    
```

### 24. 两两交换链表中的节点

给定一个链表，两两交换其中相邻的节点，并返回交换后的链表。

你不能只是单纯的改变节点内部的值，而是需要实际的进行节点交换。

```python
# 这道题有点难需要看视频理解：https://www.bilibili.com/video/BV1YT411g7br/?vd_source=e3c5eec2e1b0b0aae696b27006246ee6
# Goodnote上有图解！

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def swapPairs(self, head: ListNode) -> ListNode:
        dummy = ListNode(next=head) # 
        pre = dummy
        
        # 必须有pre的下一个和下下个才能交换，否则说明已经交换结束了
        while pre.next and pre.next.next:
            cur = pre.next
            post = pre.next.next
            
            # pre，cur，post对应最左，中间的，最右边的节点 
            cur.next = post.next
            post.next = cur
            pre.next = post
            
            # 定义next pre
            pre = pre.next.next
        return dummy.next
```

### 160.链表相交
给你两个单链表的头节点 headA 和 headB ，请你找出并返回两个单链表相交的起始节点。如果两个链表没有交点，返回 null 。

```python
# Sample Answer 1 (difficult to understand)
class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
        """
        根据快慢法则，走的快的一定会追上走得慢的。
        在这道题里，有的链表短，他走完了就去走另一条链表，我们可以理解为走的快的指针。

        那么，只要其中一个链表走完了，就去走另一条链表的路。如果有交点，他们最终一定会在同一个
        位置相遇
        """
        if headA is None or headB is None: 
            return None
        cur_a, cur_b = headA, headB     # 用两个指针代替a和b


        while cur_a != cur_b:
            # 如果cur_a 还存在， cur_a = cur_a.next, 如果cur_a不存在，return headB
            cur_a = cur_a.next if cur_a else headB      # 如果a走完了，那么就切换到b走
            cur_b = cur_b.next if cur_b else headA      # 同理，b走完了就切换到a

        return cur_a
    
        # Note: In the case lists do not intersect, the pointers for A and B
        # will still line up in the 2nd iteration, just that here won't be
        # a common node down the list and both will reach their respective ends
        # at the same time. So pA will be NULL in that case.
```

```python
# Brute force method (Not suggested)
class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
        while headA is not None:
            pB = headB
            while pB is not None:
                if headA == pB:
                    return headA
                pB = pB.next
            headA = headA.next

        return None
```

### 142.环形链表II

题意： 给定一个链表，返回链表开始入环的第一个节点。 如果链表无环，则返回 null。

为了表示给定链表中的环，使用整数 pos 来表示链表尾连接到链表中的位置（索引从 0 开始）。 如果 pos 是 -1，则在该链表中没有环。

说明：不允许修改给定的链表。

```python
# Sample Answer
# 该解法有数学证明，详情见Github解析

class Solution:
    def detectCycle(self, head: ListNode) -> ListNode:
        
        # 快慢指针都从头节点开始
        slow, fast = head, head
        
        while fast and fast.next:
            slow = slow.next # 慢指针每次走一步
            fast = fast.next.next # 快指针每次走两步
            # 如果相遇(intersection)，就意味着发现了cycle
            if slow == fast:
                p = head # 重新定义指针，p从头结点开始
                q = slow # q从intersection的位置开始
                while p!=q: # 如果q和p不相遇，p和q都前进一步，若p和q相遇，则找到了循环入口
                    p = p.next 
                    q = q.next
                #你也可以return q
                return p 

        return None
```
